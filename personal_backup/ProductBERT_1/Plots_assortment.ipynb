```python
plt.xlabel('Number of products in an Assortment ')
plt.ylabel('Time took for Training ( in secs)')
plt.plot(assortment_sizes,training_time,marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.show()
```

```python
plt.xlabel('Number of products in an Assortment ')
plt.ylabel('Validation Loss')
plt.plot(assortment_sizes,validation_loss,color='red',marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.show()
```

```python
plt.xlabel('Number of products in an Assortment ')
plt.ylabel('Time took for Validation Inference ( in secs)')
plt.plot(assortment_sizes,validation_time,marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.show()
```
